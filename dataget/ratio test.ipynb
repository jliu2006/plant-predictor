```python
import sys
import os
import glob
from glob import glob
import numpy as np
import datetime
import matplotlib.pyplot as plt
import json
```

```python
def get_ratio(folder): # 0 is NDVI, 1 is EVI
    raw_ndvi = folder + '/impute_raw_ndvi.npy'
    ref_ndvi = folder + '/impute_ref.npy'
    raw = np.load(raw_ndvi)
    ref = np.load(ref_ndvi)
    if len(ref) != 12:
        print("bruh")
        return
    res = np.zeros(shape=(len(raw), 50, 50, 2))
    for step in range(len(raw)):
        ndvi_arr = raw[step]
        ndvi_ref = ref[step%11]
        for layer in range(ndvi_arr.shape[2]):
            for i in range(ndvi_arr.shape[0]):
                for j in range(ndvi_arr.shape[1]):
                    if (ndvi_arr[i][j][layer] <= 0) or (ndvi_ref[i][j][layer] <= 0):
                        res[step][i][j][layer] = 0
                    else:
                        res[step][i][j][layer] = round(ndvi_arr[i][j][layer] / ndvi_ref[i][j][layer], 2)
    nan_mask = np.isnan(res)
    num_nan_elements = np.sum(nan_mask)
    if (num_nan_elements>0):
        print("NANS: ", num_nan_elements, folder)
    resfile = folder + '/ratio_MOD13Q1.npy'
    print(resfile, res.shape)
    np.save(resfile, res)
```

```python
folders = glob('/home/fun/wildfire_data/*')

for folder in folders:
    prof = folder + '/profile.json'
    f = open(prof)
    info = json.load(f)
    end = datetime.datetime.strptime(info['end'], '%Y-%m-%d')
    cutoff_date = end.replace(year=2021, month=2, day=1)
    if type(info['info']['acres_burned']) == int:
        if (info['info']['acres_burned'] >= 3000) and (end <= cutoff_date):
            try:
                get_ratio(folder)
            except:
                continue
```
