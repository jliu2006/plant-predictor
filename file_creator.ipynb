```python
import json
import os

def create_file(json_file):
    with open(json_file) as json_file:
        data = json.load(json_file)
    for line in data:
        print(line)
        name = line['name'] + ' ' + line['date_created'][0:10]
        name = name.replace(' ', '_')
        name = name.replace('(', '')
        name = name.replace(')', '')
        
        fire_dict = {}
        fire_dict['name'] = line['name']
        fire_dict['start'] = line['date_created'][0:10]
        fire_dict['end'] = line['last_update'][0:10]
        del line['name']
        del line['date_created']
        del line['last_update']
        fire_dict['info'] = line
        
        directory = '/home/fun/wildfire_data/' + str(name)
        if os.path.exists(directory):
            print ("%s already exists" % (directory))
        else:
            os.makedirs(directory)

        filename = directory + '/profile.json'
        #if not os.path.exists(filename):
        out_file = open(filename, "w")
        json.dump(fire_dict, out_file, indent = 6)
        out_file.close()
        
    
    
```

```python
create_file('/home/fun/plant-predictor/CA_wildfires.json')
```
