```python
import sys
import os
import glob
from glob import glob
import numpy as np
import datetime
import matplotlib.pyplot as plt
import json
```

```python
def check_data(file):
    arr = np.load(file)
    res = np.zeros(shape=(2, arr.shape[1], arr.shape[2]))
    qual = arr[2, ...]
    for i in range(qual.shape[0]):
        for j in range(qual.shape[1]):
            if (qual[i][j] == 2) or (qual[i][j] == 3):
                res[0][i][j] = np.nan
                res[1][i][j] = np.nan
            else:
                res[0][i][j] = arr[0][i][j]
                res[1][i][j] = arr[1][i][j]
    
    return res
```

```python
folders = ('/home/fun/wildfire_data/*/MOD13*.npy')
for file in glob(folders):
    res = check_data(file)
    plt.imshow(res[0])
    plt.colorbar()
    plt.show()
    
    plt.imshow(res[1])
    plt.colorbar()
    plt.show()
```
